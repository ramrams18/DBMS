```python
# This code is for mongodb client connection 
# You will get connection object from this code
# This cell is taking sometime around 5-15seconds

import urllib 
from pymongo import MongoClient

# The below code is actually used when using the localhost
# import pymongo
# myclient = pymongo.MongoClient("mongodb://localhost:27017/")

def printdocs(results,title):
    print("-"* int((86 -len(title))/2),title,"-"* int((86 -len(title))/2))
    for i in results: 
        print(i,"\n")
    print("-"*86,"\n")

username = urllib.parse.quote_plus('username')
password = urllib.parse.quote_plus('pwd')
connectionstring = 'mongodb+srv://%s:%s@cluster0.gltm7.mongodb.net/?retryWrites=true&w=majority' % (username, password)
try:
    myclient = MongoClient(connectionstring)
    mydb = myclient["db2"]
    print("Connected to server")
    
except Exception:
    print("Unable to connect to the server.")
```

```python
mycol = mydb["ARTIST"]

results = mycol.find({'artworks.form' : 'sculpture'})
printdocs(results,"ARTWORKS FORM = SCULPTURE")

results = mycol.find({'artworks.price' : {"$lte" : 10000}})
printdocs(results,"ARTWORKS PRICE <= 10000")

results = mycol.find({'name': 'Manishi Dey'})
printdocs(results," NAME = MANISHI DEY")

results = mycol.find({'birthdate': '1994-01-01'})
printdocs(results , "BIRTHDAY = 1994-01-01")

print("\n  ********* end of results ******** \n")
```
